```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def logistic(r=0.5,x_m=100,x_0=10,t=1):
    return x_m/(1+(x_m/x_0-1)*np.exp(-r*t))
```

```python

para1={'r':0.5,'x_m':1000,'x_0':10}
para2={'r':2,'x_m':1000,'x_0':10}
para3={'r':5,'x_m':1000,'x_0':10}

para4={'r':0.5,'x_m':100,'x_0':10}
para5={'r':0.5,'x_m':1000,'x_0':10}
para6={'r':0.5,'x_m':10000,'x_0':10}
```

```python
t = np.linspace(0,50)
x = logistic(**para1,t=t)
fig,axs =plt.subplots(2,3,figsize=(24,16))

axs[0][0].plot(t,logistic(**para1,t=t))
axs[0][1].plot(t,logistic(**para2,t=t))
axs[0][2].plot(t,logistic(**para3,t=t))

axs[1][0].plot(t,logistic(**para4,t=t))
axs[1][1].plot(t,logistic(**para5,t=t))
axs[1][2].plot(t,logistic(**para6,t=t))

axs[0][0].set(title="r=0.5",xlabel="t",ylabel="x")
axs[0][1].set(title="r=1.0",xlabel="t",ylabel="x")
axs[0][2].set(title="r=2.0",xlabel="t",ylabel="x")

axs[1][0].set(title="x_m=100,r=0.5",xlabel="t",ylabel="x")
axs[1][1].set(title="x_m=1000,r=0.5",xlabel="t",ylabel="x")
axs[1][2].set(title="x_m=10000,r=0.5",xlabel="t",ylabel="x")
```
