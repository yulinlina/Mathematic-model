```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
M = np.array([25,200,300,2000,5000,3e4,5e4,7e4,45e4])
H =np.array([670,420,300,205,120,85,70,72,38])
x = np.log(M)
y = np.log(H)
x,y
```

```python
def output(x):
    return -x/3
def expon(x):
    return x**(-1/3)
```

```python
b = sum(y-output(x))/len(x)
y_hat = output(x)+b
b,y_hat
```

```python
plt.figure(figsize=(8,5),dpi=80)
plt.xlabel("the weight of animals under log")
plt.ylabel("the heart of animals under log ")
plt.scatter(x,y)
plt.plot(x,y_hat)
```

```python
H_hat =(np.e**b)*expon(M)
H_hat
[abs(a-b)/b for a,b in zip(H_hat,H)]
```

```python
sum([abs(a-b)/b for a,b in zip(H_hat,H)])
```

```python
plt.figure(figsize=(8,5),dpi=80)
plt.xlabel("the weight of animals ")
plt.ylabel("the heart of animals ")
plt.title("Fitting of the raw data")
plt.scatter(M,H,c='red',marker='o',alpha=0.5)
blue_line =mlines.Line2D([],[],linestyle='-',color='blue',markersize=2,label= 'fitting data')
plt.legend(handles=[blue_line])
plt.plot(M,H_hat,linestyle='-')
```
